Menjadikan semua nama column huruf kecil . 

```python
# Normalisasi nama kolom
df.columns = df.columns.str.lower()  # Mengubah menjadi lowercase
df.columns = df.columns.str.replace(' ', '_')  # Mengubah spasi menjadi underscore
df.columns = df.columns.str.replace('[^\w\s]', '')  # Menghapus karakter selain huruf dan underscore

```

```python
df.columns
```

```python
df.numberoffloors.value_counts()
```

Menganti kolum

```python
df.drop(['electricitykwhprice','electricitybaseprice','lastrefurbish','energyefficiencyclass','facilities','description','housenumber'],axis=1,inplace =True)
```

```python
# Replace NaN values based on specified conditions
df['petsallowed'] = df['petsallowed'].fillna('no')
df['interiorqual'] = df['interiorqual'].fillna('none')
df['condition'] = df['condition'].fillna('not_specified')
df['typeofflat'] = df['typeofflat'].fillna('not_specified')
df['telekomtvoffer']=df['telekomtvoffer'].fillna('NONE')
df['telekomhybriduploadspeed']=df['telekomhybriduploadspeed'].fillna(0)
df['noparkspaces']=df['noparkspaces'].fillna(0)
df['telekomuploadspeed']=df['telekomuploadspeed'].fillna(0)
df['yearconstructedrange']=df['yearconstructedrange'].fillna(8.0)
# Replace missing values in heatingcost with the mean
df['heatingcosts'] = df['heatingcosts'].fillna(df['heatingcosts'].mean())
df['thermalchar'] = df['thermalchar'].fillna(df['thermalchar'].mean())
  

df['numberoffloors'] = df['numberoffloors'].fillna(method='bfill')
# Handle missing values in servicecharge
df['servicecharge'] = df['servicecharge'].fillna(0)  # You can use mean() or other strategies instead of 0

# Handle missing values in pricetrend
df['pricetrend'] = df['pricetrend'].fillna(df['pricetrend'].mean())  # You can use other strategies as needed

# Handle missing values in heatingtyped
df['heatingtype'] = df['heatingtype'].fillna('not_specified')  # Replace with a suitable category
# Handle missing values in totalrent
df['totalrent'] = df['totalrent'].fillna(df['totalrent'].mean())  # You can use other strategies as needed

# Handle missing values in yearconstructed
df['yearconstructed'] = df['yearconstructed'].fillna(df['yearconstructed'].median())  # You can use other strategies as needed

# Handle missing values in firingtypes
df['firingtypes'] = df['firingtypes'].fillna('not_specified')  # Replace with a suitable category

# Handle missing values in streetplain
df['streetplain'] = df['streetplain'].fillna('not_specified')  # Replace with a suitable category

# Handle missing values in floor
df['floor'] = df['floor'].fillna(df['floor'].median())  # You can use other strategies as needed
```
